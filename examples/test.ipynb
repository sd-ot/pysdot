```python
from pysdot import PowerDiagram
import numpy as np

pd = PowerDiagram(np.random.rand(11, 2))
pd.display_jupyter()
```

```python
positions = np.random.rand(10, 2)
weights = np.ones(positions.shape[0])

import ipywidgets
nsl0 = ipywidgets.widgets.FloatSlider(min=1-0.1,max=1+0.1,step=0.001,value=1.0,layout=ipywidgets.widgets.Layout(width='100%'))
nsl1 = ipywidgets.widgets.FloatSlider(min=1-0.1,max=1+0.1,step=0.001,value=1.0,layout=ipywidgets.widgets.Layout(width='100%'))
@ipywidgets.interact(w0=nsl0, w1=nsl1)
def disp(w0, w1):
    weights[ 0 ] = w0
    weights[ 1 ] = w1

    pd = PowerDiagram(positions, weights)
    return pd.display_jupyter()
```

```python
from scipy.sparse.linalg import spsolve
from scipy.sparse import csr_matrix

mvs = pd.der_integrals_wrt_weights()
N = pd.positions.shape[0]
A = csr_matrix((mvs.m_values, mvs.m_columns, mvs.m_offsets))
V = mvs.v_values - np.ones(N) / N

A[0, 0] *= 2

X = spsolve(A, V)
pd.weights -= X

pd.display_jupyter()
```
